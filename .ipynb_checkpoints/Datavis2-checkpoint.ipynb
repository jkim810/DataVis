```python
import plotly.plotly as py
import pandas as pd
```

```python
df = pd.read_csv('airline.csv')
```

```python
df
```

```python
for col in df.columns:
    df[col] = df[col].astype(str)

scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
        [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
```

```python
for month in df[' month']:
    if str(month) == '2':
        print(df['arr_flights'])
```

```python
df[' month']
```

```python
df['text'] = 'Arriving Flights' + df['arr_flights']

#df1 = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv')
```

```python
a = str(df['airport_name']).split()
l = []
for e in a:
    if len(e) == 3:
        if e[2] == ':':
            l.append(e[0:2])
        
```

```python
data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        #locations = df['code'],
        locations = l,
        #z = df['total exports'].astype(float),
        z = df['arr_flights'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Millions USD")
        ) ]
```

```python
layout = dict(
        title = '2011 US Agriculture Exports by State<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
```

```python
fig = dict(data=data, layout=layout)
py.iplot(fig,filename='d3-cloropleth-map')
```
