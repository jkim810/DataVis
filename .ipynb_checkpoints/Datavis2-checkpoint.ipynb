```python
df = pd.read_csv('airline.csv')

for col in df.columns:
    df[col] = df[col].astype(str)

l = []
a = df['airport_name']

for e in a:
    tmp = e.split(' ')
    for i in tmp:
        if ':' in i:
            l.append(i[0:2])

df['state'] = pd.Series(v for v in l)
df.drop(columns = ['airport', 'airport_name', 'carrier','carrier_name'])
```

```python
# Pie Chart for delay Cause
# Calculate Occurance of Carrier delays, weather delays, secuirty delays, and late aircraft delays

dc, sum_dc = pd.Series(), pd.Series()
dc['carrier_ct'] = df['carrier_ct'].astype(float)
dc['weather_ct'] = df['weather_ct'].astype(float)
dc['security_ct'] = df['security_ct'].astype(float)
dc['late_aircraft_ct'] = df['late_aircraft_ct'].astype(float)

tmp = 0.0
for i in dc['carrier_ct']:
    if i == i:
        tmp = tmp + i
sum_dc['carrier_ct'] = tmp

tmp = 0.0
for i in dc['weather_ct']:
    if i == i:
        tmp = tmp + i
sum_dc['weather_ct'] = tmp

tmp = 0.0
for i in dc['security_ct']:
    if i == i:
        tmp = tmp + i
sum_dc['security_ct'] = tmp

tmp = 0.0
for i in dc['late_aircraft_ct']:
    if i == i:
        tmp = tmp + i
sum_dc['late_aircraft_ct'] = tmp
```

```python
labels = ['Carrier Delays','Weather Delays','Security Delays','Late Aircraft Arrival Delays']
values = list(sum_dc)
colors = ['#FEBFB3', '#E1396C', '#96D38C', '#D0F9B1']
```

```python
trace = go.Pie(labels=labels, values=values,
               hoverinfo='label+percent', textinfo='label', 
               textfont=dict(size=20),
               marker=dict(colors=colors, 
                           line=dict(color='#000000', width=2)),
               
              )
data=go.Data([trace])
layout=go.Layout(title="Causes for Flight Delays",showlegend=True)
figure=go.Figure(data=data,layout=layout)
py.iplot(figure, filename='Delay Causes')
```
